```python
import pandas as pd
import numpy as np
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import torch
import torch.nn as nn
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from transformers import BitsAndBytesConfig
from transformers import AutoConfig

train_data = pd.read_csv("/content/drive/MyDrive/train.csv")
# Define emotion columns
emotion_columns = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'love', 'optimism', 'pessimism', 'sadness', 'surprise', 'trust']

# Quantization configuration
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=False
)
# Step 4: Initialize tokenizer and set padding token
model_name = "Qwen/Qwen2-1.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'right'

train_df, val_df = train_test_split(train_data, test_size=0.2, random_state=42)
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)

print("Train dataset size:", len(train_dataset))
print("Val dataset size:", len(val_dataset))



```

```python
def tokenize_and_format(examples):
    # Ensure padding to max_length
    tokenized = tokenizer(
        examples['Tweet'],
        padding='max_length',  # Force padding to max_length
        truncation=True,
        max_length=128,
        return_tensors=None    # Don't return tensors yet
    )

    # Format labels
    labels = torch.zeros((len(examples['Tweet']), len(emotion_columns)))
    for i, example in enumerate(examples['Tweet']):
        for j, emotion in enumerate(emotion_columns):
            labels[i][j] = examples[emotion][i]

    tokenized['labels'] = labels.tolist()

    # Verify lengths
    max_len = max(len(ids) for ids in tokenized['input_ids'])
    for i in range(len(tokenized['input_ids'])):
        if len(tokenized['input_ids'][i]) < max_len:
            # Pad input_ids
            padding_length = max_len - len(tokenized['input_ids'][i])
            tokenized['input_ids'][i].extend([tokenizer.pad_token_id] * padding_length)
            # Pad attention_mask
            tokenized['attention_mask'][i].extend([0] * padding_length)

    return tokenized

```

```python
# Process datasets with the updated function
tokenized_train = train_dataset.map(
    tokenize_and_format,
    batched=True,
    remove_columns=train_dataset.column_names,
    batch_size=8  # Process in smaller batches
)
tokenized_val = val_dataset.map(
    tokenize_and_format,
    batched=True,
    remove_columns=val_dataset.column_names,
    batch_size=8
)

# Set format for pytorch
tokenized_train.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])
tokenized_val.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])


# Step 7: Calculate class weights
train_labels = np.array([sample['labels'] for sample in tokenized_train])
label_counts = np.sum(train_labels, axis=0)
total_samples = len(tokenized_train)
class_weights = torch.tensor(total_samples / (len(emotion_columns) * label_counts), dtype=torch.float)
```

```python
# Add data collator to trainer initialization
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# Step 8: Define model class
class Qwen2MultiLabelClassification(nn.Module):
    def __init__(self, num_labels, class_weights):
        super().__init__()
        self.num_labels = num_labels
        self.class_weights = class_weights

        # Configure model config first
        model_config = AutoConfig.from_pretrained(model_name)
        model_config.pad_token_id = tokenizer.pad_token_id
        model_config.num_labels = num_labels  # Set num_labels in config
        model_config.problem_type = "multi_label_classification"

        # Load model with config
        self.model = AutoModelForSequenceClassification.from_pretrained(
            model_name,
            config=model_config,
            quantization_config=bnb_config,
        )

        # Prepare for QLoRA
        self.model = prepare_model_for_kbit_training(self.model)

        # Configure LoRA
        lora_config = LoraConfig(
            r=32,
            lora_alpha=64,
            target_modules=[
                "q_proj",
                "k_proj",
                "v_proj",
                "o_proj",
                "gate_proj",
                "up_proj",
                "down_proj",
            ],
            lora_dropout=0.05,
            bias="none",
            task_type="SEQUENCE_CLASSIFICATION"
        )

        self.model = get_peft_model(self.model, lora_config)
        self.loss_fct = nn.BCEWithLogitsLoss(pos_weight=class_weights)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        loss = None
        if labels is not None:
            loss = self.loss_fct(logits, labels.float())

        return {'loss': loss, 'logits': logits} if loss is not None else {'logits': logits}
```

```python
def compute_metrics(pred):
    predictions = (pred.predictions > 0).astype(np.int32)
    return {
        'macro_f1': f1_score(pred.label_ids, predictions, average='macro'),
        'micro_f1': f1_score(pred.label_ids, predictions, average='micro'),
        'weighted_f1': f1_score(pred.label_ids, predictions, average='weighted')
    }
```

```python
training_args = TrainingArguments(
    output_dir='./results_qwen2_qlora',
    num_train_epochs=10,
    per_device_train_batch_size=2,  # Reduced from 4
    per_device_eval_batch_size=4,   # Reduced from 8
    gradient_accumulation_steps=16,  # Increased from 8
    learning_rate=1e-4,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1",
    greater_is_better=True,
    warmup_ratio=0.2,
    logging_steps=100,
    fp16=True,
    remove_unused_columns=False,
    save_total_limit=2,
    report_to="wandb"
)
```

```python
# Step 11: Initialize model and trainer
model = Qwen2MultiLabelClassification(  # Changed from QwenMultiLabelClassification
    num_labels=len(emotion_columns),
    class_weights=class_weights
)
```

```python
# Update trainer initialization
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    compute_metrics=compute_metrics,
    data_collator=data_collator,  # Add data collator
)
```

```python
# Step 12: Train the model
trainer.train()
trainer.save_model("./final_qwen2_model")
```

```python
# Step 13: Generate predictions for test set
test_data = pd.read_csv("/content/drive/MyDrive/test.csv")
# Step 13: Generate predictions for test set
test_dataset = Dataset.from_pandas(test_data)  # Convert DataFrame to Dataset

# Now we can use column_names
tokenized_test = test_dataset.map(
    lambda x: tokenizer(
        x['Tweet'],
        padding="max_length",
        truncation=True,
        max_length=128
    ),
    batched=True,
    remove_columns=test_dataset.column_names  # Now this will work
)
tokenized_test.set_format(type='torch', columns=['input_ids', 'attention_mask'])

# Make predictions
predictions = trainer.predict(tokenized_test)
binary_predictions = (predictions.predictions > 0).astype(int)

# Create submission DataFrame
submission = pd.DataFrame(binary_predictions, columns=emotion_columns)
submission['ID'] = test_data['ID']
submission = submission[['ID'] + emotion_columns]
submission.to_csv('qwen2_predictions.csv', index=False)

print("Training and prediction completed. Results saved to 'qwen2_predictions.csv'")
```
